```python
import torch
import torchvision.models as models
import torch.onnx

# Загрузка и запуск предобученной модели ResNet-50
model = models.resnet50(pretrained=True)
input_data = torch.randn(1, 3, 224, 224)
output = model(input_data)
print(output)

# Конвертация модели в формат ONNX
!pip install onnx onnxruntime
import onnx
import onnxruntime
onnx_model_path = "resnet50_no_bn_no_dropout.onnx"
torch.onnx.export(model, input_data, onnx_model_path)

# Загрузка и запуск модели в ONNX
onnx_model = onnx.load(onnx_model_path)
ort_session = onnxruntime.InferenceSession(onnx_model_path)
ort_inputs = {ort_session.get_inputs()[0].name: input_data.numpy()}
ort_outs = ort_session.run(None, ort_inputs)
print("Результат запуска модели Resnet50 в ONNX:", ort_outs)
```

```python
#Конвертация в PyTorch и повторный запуск модели
!pip install onnx2torch
from onnx2torch import convert
torch_model = convert(onnx_model_path)
output = torch_model(input_data)
print(output)
```
